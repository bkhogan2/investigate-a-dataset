I've chosen to investigate the relationship between hits and wins. Let's get started!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('data/core/People.csv')
teams = pd.read_csv('data/core/Teams.csv')
teams = teams[teams['G'] >= 160]

teams[['H', 'yearID','G', 'R']]
```

To start, I'll just go right for it. The next plot shows the relationship between hits and runs for teams with games played greater than or equal to 60.

```python
hits_wins = teams[['H', 'W']]
hits = teams['H']
wins = teams['W']

plt.scatter(hits, wins, color="black")
plt.xlabel("Hits")
plt.ylabel("Wins")
plt.title("Correlation between hits and wins")
```

```python
hits_r = (hits - hits.mean()) / hits.std(ddof=0) 
wins_r = (wins - wins.mean()) / wins.std(ddof=0)

r = (hits_r * wins_r).mean()
print(r)

plt.plot(hits, hits*r, color='red')
```
